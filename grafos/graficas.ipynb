```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos_profundidad = pd.read_csv('resultados_spanning_profundidad.csv')
```

```python
x=datos_profundidad['Nodos'].values
y=datos_profundidad['Tiempo_Segundos'].values

coeficientes = np.polyfit(x, y, 3)

funcion_cubica = np.poly1d(coeficientes)
x_curva = np.linspace(x.min(), x.max(), 500)
y_curva = funcion_cubica(x_curva)



plt.figure(figsize=(10, 6)) 

plt.plot(x_curva, y_curva, color='blue', linewidth=2, label='Ajuste cubico')

plt.scatter(x, y, color='red', label='Datos Experimentales (Reales)')
```
